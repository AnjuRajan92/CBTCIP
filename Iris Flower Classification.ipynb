```python
# Import required libraries
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
```

```python
#Load the dataset from scikitlearn
from sklearn.datasets import load_iris
df=load_iris()
type(df)
```

```python
df.keys()
```

```python
# feature names(variable names)
df['feature_names']
```

```python
#target names
df['target_names']
```

```python
df1=pd.DataFrame(data=df['data'],columns=df['feature_names'])
```

```python
df1['Iris_category']=df['target']
```

```python
df1
```

```python
df1['Species'] = df1['Iris_category'].apply(lambda x: 'sentosa' if x == 0 else ('versicolor' if x == 1 else 'virginica'))
#iris_data['Species'] = iris['target']
df1
```

```python
df1.shape
```

```python
df1.info()
```

```python
df1.isnull().sum()
```

# Data Cleaning and Preprocessing

```python
df1['Species'].value_counts()
```

```python
df1.drop(columns='Iris_category',axis=1,inplace=True)
df1.head(3)
```

```python
df1[df1.duplicated()]
```

```python
#Drop duplicate value
df1.drop_duplicates(inplace=True)
df1[df1.duplicated()]
```

```python
#Label encoding
Label=LabelEncoder()
df1['Species']=Label.fit_transform(df1['Species'])
df1.head()
```

```python
df1.tail(3)
```

```python
df1.shape
```

# Splitting the dataset into training set and testing set

```python
x=df1.drop(columns=['Species'])
x.head(3)
```

```python
y=df1['Species']
y.head(3)
```

```python
#Split into training and testing
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30)
```

# Logistic Regression

```python
LR_model=LogisticRegression()
LR_model.fit(x_train,y_train)
```

```python
y_pred=LR_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
print("Accuracy score:",accuracy_score(y_test,y_pred))
print("Classification report:",classification_report(y_test,y_pred))
```

# Random Forest Classifier

```python
RF_model=RandomForestClassifier()
RF_model.fit(x_train,y_train)
```

```python
y_pred1=RF_model.predict(x_test)
```

```python
print("Accuracy score:",accuracy_score(y_test,y_pred1))
print("Classification report:",classification_report(y_test,y_pred1))
```
